```python
import pandas as pd

data1 ={
    "id":[0,1,2,3,4],
    "location":["USA","USA","USA","USA","USA"],
    "age":[24,31,29,33,36],
    "gender":["M","F","F","M","F"],
    "is_senior":[False,True,False,False,True]
}

data2 ={
    "employee_id":[0,1,2,3,4],
    "age":[24,31,29,33,36],
    "gender":["M","F","F","M","F"],
    "popularity":[6,4,0,7,6]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
# Merge two dataframes

merged_df = pd.merge(df1, df2, on=["age","gender"],how="outer")
print(merged_df)
```

```python
# Find average popularity as per location

avg_popularity = merged_df.groupby("location")["popularity"].mean()
print(avg_popularity)
```

## Matplotlib (Graph)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://itv-contentbucket.s3.ap-south-1.amazonaws.com/Exams/AWP/Matplotlib/historical_automobile_sales.csv"
df = pd.read_csv(url)

print(df.head())
```

```python
# 1) Develop a Line chart using the functionality of pandas to show how. automobile sales fluctuate from year to year

yearly_sales = df.groupby('Year')['Automobile_Sales'].sum()
yearly_sales.plot(kind='line', figsize=(10,5), marker='o',color='red',mfc='g',mec='b')
plt.title("Yearly Automobile Sales")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
# 2) Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?

rec = df[df['Recession'] == 1]
vehicle_trend = rec.groupby(['Year','Vehicle_Type'])['Automobile_Sales'].sum().unstack()

vehicle_trend.plot()
plt.title("Sales Trend by Vehicle Type")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
# 3) Use the Matplotlib Library to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.

recession = df[df['Recession'] == 1]
non_recession = df[df['Recession'] == 0]

rec_sales = recession.groupby('Vehicle_Type')['Automobile_Sales'].mean()
nonrec_sales = non_recession.groupby('Vehicle_Type')['Automobile_Sales'].mean()

rec_sales.plot(kind='bar')
plt.title("Vehicle Sales during Recession")
plt.xlabel("Vehicle Type")
plt.ylabel("Average Sales")
plt.show()

nonrec_sales.plot(kind='bar')
plt.title("Vehicle Sales during Non-Recession")
plt.xlabel("Vehicle Type")
plt.ylabel("Average Sales")
plt.show()
```

```python
# 4) Use the functionality of Matplotlib to develop a scatter plot to identify the correlation between average vehicle price relate to the sales volume during recessions.

plt.scatter(recession['Price'], recession['Automobile_Sales'])
plt.title("Price vs Sales during Recession")
plt.xlabel("Average Vehicle Price")
plt.ylabel("Sales")
plt.show()
```

```python
# 5) Create a pie chart to display the portion of advertising expenditure of. Automotive during recession and non-recession periods.

ad_exp = df.groupby('Recession')['Advertising_Expenditure'].sum()

plt.pie(ad_exp, labels=['Non-Recession', 'Recession'], autopct='%1.1f%%')
plt.title("Advertising Expenditure Distribution")
plt.show()
```
